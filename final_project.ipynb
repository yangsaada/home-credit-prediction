Since the column are all categorical. We can convert them to numerical by doing one hot encoding for column with category more than 2 and by label encoding for column with category only 2 (0 for no, 1 for yes).

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
# Label Encoding
le = LabelEncoder()
le_count = 0

for col in df_train.select_dtypes('object'):
    if df_train[col].nunique() == 2:
        # Fit to train data
        le.fit(df_train[col])
        # Transform to both train and test data
        df_train[col] = le.transform(df_train[col])
        df_test[col] = le.transform(df_test[col])
        # Count up total transformed column
        le_count = le_count + 1

print("Total columns encoded: {}".format(le_count))
```

```python
enc = OneHotEncoder(sparse=False,handle_unknown='ignore')
enc_count = 0

for col in df_train.select_dtypes('object'):
    # Fit to train data
    enc.fit(df_train[[col]])
    # Transform data train
    data_train = enc.transform(df_train[[col]])
    columns_name = enc.get_feature_names_out()
    temp_df = pd.DataFrame(data=data_train, columns=columns_name)
    df_train.drop(columns=col, axis=1, inplace=True)
    df_train = pd.concat([df_train, temp_df], axis=1)
    # Transform data test
    data_test = enc.transform(df_test[[col]])
    columns_name = enc.get_feature_names_out()
    temp_df = pd.DataFrame(data=data_test, columns=columns_name)
    df_test.drop(columns=col, axis=1, inplace=True)
    df_test = pd.concat([df_test, temp_df], axis=1)
    # Count up total transformed column
    enc_count = enc_count + 1

print("Total columns encoded: {}".format(enc_count))
```

The max is positive? This is wierd. Let's see the histogram

```python
sns.histplot(data=df_train, x="DAYS_EMPLOYED")
```

This positive number is surely are anomalies. Let's remove their value by replacing with null value.
